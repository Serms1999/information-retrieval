```python

import os
import pandas as pd


# Si el archivo está en un directorio diferente, proporciona la ruta completa
df_adult = pd.read_csv(os.path.join(os.pardir,"data","adult","adult.csv"), header = None, index_col = None ,names = [
    "age",
    "workclass",
    "fnlwgt",
    "education",
    "education-num",
    "marital-status",
    "occupation",
    
    "relationship",
    "race",
    "sex",
    "capital-gain",
    "capital-loss",
    "hours-per-week",
    "native-country", "class"
])


```

```python
df_adult
```

```python
print(df_adult.columns)
```

```python
# Strip leading and trailing spaces
df_adult['class'] = df_adult['class'].str.strip()

# Map 'label_value' column to binary
df_adult['class'] = df_adult['class'].map({'<=50K': 0, '>50K': 1})
```

```python
df_adult
```

```python
from sklearn.preprocessing import OneHotEncoder

# Select the categorical columns
categorical_cols = df_adult.select_dtypes(include=['object']).columns

# Instantiate the OneHotEncoder
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')

# Fit and transform the categorical columns
one_hot_encoded_cols = encoder.fit_transform(df_adult[categorical_cols])

# Create a DataFrame from the one-hot encoded columns
df_encoded = pd.DataFrame(one_hot_encoded_cols, columns=encoder.get_feature_names_out(categorical_cols))

# Drop the original categorical columns from df_adult
df_adult = df_adult.drop(columns=categorical_cols)

# Concatenate df_adult and df_encoded
df_adult = pd.concat([df_adult, df_encoded], axis=1)
```

```python
df_adult
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X = df_adult.drop('class', axis=1)
y = df_adult['class']

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create a logistic regression model
model = LogisticRegression()

# Train the model
model.fit(X_train, y_train)
```

```python
X_train
```

```python
# Make predictions on the test data
y_pred = model.predict(X_test)
```

```python
# Convert y_pred to a pandas Series
y_pred_series = pd.Series(y_pred, index=X_test.index, name='label_value')

# Join y_test, y_pred_series with X_test
df_test = pd.concat([X_test, y_test, y_pred_series], axis=1)
```

```python
# Rename the 'class' column to 'score' and 'label_value' to 'predicted'
df_test = df_test.rename(columns={'class': 'label_value', 'label_value': 'score'})
```

```python
df_test
```

```python
# Create a new list of column names with 'score' and 'label_value' first
cols = ['class', 'label_value'] + [col for col in df_test.columns if col not in ['class', 'label_value']]

# Create a new DataFrame with the columns reorganized
df_test_reorganized = df_test[cols]
```

```python
df_test_reorganized
```

```python
df_test_reorganized = df_test_reorganized.rename(columns={'class': 'score'})
```

```python
df_test_reorganized
```

```python
df_test_reorganized.to_csv('df_adult_Fairness.csv', index=False)
```
